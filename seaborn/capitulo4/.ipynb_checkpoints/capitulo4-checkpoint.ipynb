# Nesse capítulo vou explorar algumas customizações nos gráficos do seaborn, como:
1. Mudança de style(Fundo do gráfico)
2. Mudança de palette(Cores do gráfico)
3. Mudança de context(Maneira de como o gráfico é apresentado)
4. Adição de título
5. Mudança os labels do eixo x e eixo y

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
survey_data = pd.read_csv("../dados/young-people-survey-responses.csv")
```

```python
survey_data.head()
```

## Para mudar o estilo, devemos chamar a função sns.set_style() com o nome do estilo. Existem os estilos:
1. white
2. dark
3. whitegrid
4. darkgrid
5. ticks


```python
sns.set_style("whitegrid")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_style("white")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_style("dark")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_style("ticks")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

## Para mudar as cores que aparecem nos gráficos, existem diversos padrões de cores que podemos usar. Devemos chamar a função sns.set_palette(). Algumas são:


## Ainda podemos criar a nossa própria palette criando uma lista com as cores



## Testar essas palettes: 'deep', 'muted', 'pastel', 'bright', 'dark','colorblind'

```python
sns.set_style("whitegrid")
sns.set_palette("PuRd")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("RdBu")
sns.catplot(x="Internet usage", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette(["#39A7D0", "#36ADA4"])
sns.catplot(x="Gender", y="Age", 
            data=survey_data, kind="box")
plt.show()
```

## Para mudar a forma de como o gráfico é apresentado, devemos chamar a função sns.set_context. Existem quatro tipos:
1. paper
2. notebook
3. talk
4. poster



```python
sns.set_context("paper")
sns.catplot(x="Gender", y="Siblings",
            data=survey_data, kind="bar")
plt.show()
```

```python
sns.set_context("notebook")
sns.catplot(x="Gender", y="Siblings",
            data=survey_data, kind="bar")
plt.show()
```

```python
sns.set_context("talk")
sns.catplot(x="Gender", y="Siblings",
            data=survey_data, kind="bar")
plt.show()
```

```python
sns.set_context("poster")
sns.catplot(x="Gender", y="Siblings",
            data=survey_data, kind="bar")
plt.show()
```

## Existem duas formas de colocar um título para o gráfico que depende que qual função estamos utilizando:

## As formas de colocas os títulos para esses dois cenários são:


```python
mpg = pd.read_csv("../dados/mpg.csv")
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```

## Para modificar as labels do eixo x e eixo y devemos chamar a função set(xlabel = "título do eixo x", ylabel="Título do eixo y")


```python
g = sns.lineplot(x="model_year", y="horsepower", 
                 data=mpg,
                 hue="origin")
g.set_title("Average MPG Over Time")
g.set(xlabel="Car Model Year", ylabel="Average MPG")
plt.show()
```

## Para rotacionar as labels do eixo x, devemos chamar xticks do matplotlib e fornecer o ângulo de rotação:


```python
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)
plt.xticks(rotation=90)
plt.show()
```
