# O Seaborn é uma bibliotecas mais famosas de visualização de dados, vou explorar diversas visualizações bem como quando usar cada uma e as diferenças.
## As visualizações que vamos explorar nesse capítulo são:
1. Gráficos de pontos
2. Gráficos de barras
3. Como separar as observações por cor

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
student_data = pd.read_csv("../dados/student-alcohol-consumption.csv")
```

### Gráficos de pontos padrão

```python
sns.scatterplot(x="absences", y="G3", data=student_data)
plt.show()
```

### Gráfico de pontos, só que mudando a cor do pontos com base em uma variável categórica

```python
sns.scatterplot(x="absences", y="G3", data=student_data, hue="location")
plt.show()
```

## Gráfico de Barras padrão

```python
sns.countplot(x="school", data=student_data)
plt.show()
```

## Gráfico de Barras, só que separando por mais uma variável categórica e escolhendo as cores manualmente com o argumento palette

```python
palette_colors = {"Rural": "green", "Urban": "blue"}
sns.countplot(x="school", data=student_data, hue="location", palette=palette_colors)
plt.show()
```
