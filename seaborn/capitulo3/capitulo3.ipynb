1. Gráfico de barras com uma variável numérica no eixo y
2. Box plots
3. Gráficos de pontos com o eixo x sendo uma variável categórica
4. Gráfico de barra usual

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
survey_data = pd.read_csv("../dados/young-people-survey-responses.csv")
student_data = pd.read_csv("../dados/student-alcohol-consumption.csv")
```

## A grande diferença entre o countplot e catplot é que com o catplot dá pra criar sub-gráficos com os argumentos col e row, da mesma forma que o relplot



## Gráfico de barras usual, só que usando a função catplot, agora temos que usar o argumento kind=count

```python
sns.catplot(x="Internet usage", data=survey_data, kind="count")
plt.show()
```

## Gráfico de barras usual, mudando a orientação das barras usando o argumento y ao invés de x

```python
sns.catplot(y="Internet usage", data=survey_data, kind="count")
plt.show()
```

## Gráfico de barras, só que criando sub-gráficos com o argumento col baseado em uma variável categórica

```python
sns.catplot(y="Internet usage", data=survey_data, kind="count",col="Gender")
plt.show()
```

```python
survey_data["internet_usage_bool"] = survey_data["Internet usage"]=="few hours a day"
```

## Gráfico de barras, só que usando uma variável numérica ou booleana no eixo y. Nesse caso booleana, devemos usar o argumento kind="bar"

```python
sns.catplot(x="Gender", y="internet_usage_bool", data=survey_data, kind="bar")
plt.show()
```

## Gráfico de barras usando uma variável numérica no eixo y, aqui a média é usada como altura na barra e um intervalo de confiança é usado

```python
sns.catplot(x="study_time", y="G3", data=student_data, kind="bar")
plt.show()
```

## Gráfico de barras usando uma variável numérica no eixo y, e colocando uma ordem específica do eixo x

```python
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order)
plt.show()
```

## Gráfico de barras usando uma variável numérica no eixo y, e tirando o intervalo de confiança

```python
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order,
            ci=None)
plt.show()
```

## Criando um boxplot com uma variável categórica no eixo x e uma variável numérica no eixo y

```python
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]
sns.catplot(x="study_time", y="G3", data=student_data, kind="box", order=study_time_order)
plt.show()
```

## Criando um boxplot com uma variável categórica no eixo x e uma variável numérica no eixo y. Mudando a cor dos boxplots de acordo com base em outra variável categórica

```python
sns.catplot(x="internet", y="G3", data=student_data, hue="location", kind="box", sym="")
plt.show()
```

## O boxplot usa 1.5*IQ para determinar se um valor é outlier ou não, para mudarmos o 1.5, devemos usar o argumento whis. Quanto menor, mais outliers serão mostrados, quanto maior, menos outliers serão mostrados.

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box", whis=0.5)
plt.show()
```

## Se eu forneço uma lista com dois valores entre 0 e 100 no argumentos whis, ele vai considerar esses valores como os percentis para mostrar os outliers.

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5, 95])
plt.show()
```

## Nesse caso, os limites são o mínimo e máximo do conjunto de dados, logo não haverá nenhum outlier

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])
plt.show()
```

## Point plots são muito parecido com o gráfico de barras com uma variável numérica no eixo y, porém agora só mostramos o ponto e não a barra


## A grande diferença de point plots para line plots(quando temos vários valores de y paraa um mesmo valor de x) é que no gráfico de linhas a variável do eixo x é numérica e no point plot é categórica


## Para o gráfico point plot, devemos colocar o argumento kind="point"

```python
sns.catplot(x="famrel", y="absences", data=student_data, kind="point")
plt.show()
```

## Para adicionar a barra no intervalo de confiança, devemos colocar a largura dessa bara no argumento capsize

```python
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2)
plt.show()
```

## Para tirar as ligações entre os pontos, devemos colocar o argumento join=False

```python
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2,
            join=False)
plt.show()
```

## Para mudar a cor com base em uma variável categórica, podemos usar o argumento hue como de costume

```python
sns.catplot(x="romantic", y="absences", data=student_data, hue="school", kind="point")
plt.show()
```

## Para tirar o intervalo de confiança, devemos colocar o argumento ci=None

```python
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None)
plt.show()
```

## Para usarmos a mediana ao invés da média como medida de resumo nos pontos, devemos usar o argumento estimator

```python
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None, estimator=median)
plt.show()
```
