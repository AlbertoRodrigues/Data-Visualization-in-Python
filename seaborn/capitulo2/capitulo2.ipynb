# Nesse capítulo, vamos explorar:
1. Gráfico de pontos usando relplot
2. Gráfico de linhas usando relplot
3. Argumentos adicionais do relplot, como style, size, markers, dashes e intervalos de confiança

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
```

```python
student_data = pd.read_csv("../dados/student-alcohol-consumption.csv")
```

## A função do seaborn relplot pode criar gráficos de pontos ou linhas, a grande vantagem é poder criar diversos subplots em um mesmo gráfico, o que não é possível usando scatterplot.
### A sintaxe é bem parecida, o que é muda é o argumento kind para escolher o tipo do gráfico:


## Nesse caso, plotei vários sub-gráficos horizontalmente usando o argumento col com base na variável study_time. Também é possível colocar em várias linhas usando o argumento row.

```python
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
            col="study_time")
plt.show()
```

```python
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
            row="study_time")
plt.show()
```

## Podemos mudar a ordem de aparição com o argumento col_order ou row_order

```python
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter",
            col="schoolsup",
            col_order=["yes", "no"])
plt.show()
```

## Também podemos misturar sub gráficos em linhas e colunas usando duas variável categóricas diferentes

```python
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])
plt.show()
```

## Podemos especificar quantas colunas queremos com o argumento col_wrap

```python
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
            col="study_time",
           col_wrap=2)
plt.show()
```

```python
mpg = pd.read_csv("../dados/mpg.csv")
```

## Podemos mudar o tamanho dos pontos com base em uma variável categórica usando o argumento size

```python
sns.relplot(x="horsepower", y="mpg", data=mpg, kind="scatter", size="cylinders")
plt.show()
```

## Mudando a cor e o tamanho com base na mesma variável categórica

```python
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",
            hue="cylinders")
plt.show()
```

## Mudando o estilo dos pontos com base em uma variável categórica

```python
sns.relplot(x="acceleration", y="mpg", kind="scatter", style="origin", hue="origin", data=mpg)
plt.show()
```

## Mudando a transparência dos pontos com o argumento alpha, com alpha=0 sendo totalmente transparente e alpha=1 sendo o mais "cheio" possível

```python
sns.relplot(x="horsepower", y="mpg", data=mpg, kind="scatter", alpha=0.4)
plt.show()
```

## Existem dois tipos de gráficos de linhas:
1. Existe somente um valor de y para cada valor de x
2. Existem múltiplos valores de x para um mesmo valor de y
### No primeiro caso o gráfico seria dessa forma:


### No segundo caso o gráfico pega por padrão a média de y para o mesmo conjunto de valores de x:



## Por padrão, o gráfico constrói um intervalo de confiança de 95% para a média para cada valor do eixo x

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line")
plt.show()
```

## Usando a mediana ao invés da média

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line", estimator=median)
plt.show()
```

## Usando um desvio padrão ao invés do intervalo de confiança

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line", ci="sd")
plt.show()
```

## Tirando o intervalo de confiança

```python
sns.relplot(x="model_year", y="horsepower", data=mpg, kind="line", ci=None)
plt.show()
```

## Mudando o estilo e cor das linhas com base em uma variável categórica

```python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, hue="origin",
            style="origin")
plt.show()
```

## Desse jeito acrescenta pontos e muda o estilo deles e não das linhas

```python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",
            dashes=False,
            markers=True)
plt.show()
```
